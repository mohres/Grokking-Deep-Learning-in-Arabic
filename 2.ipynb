<h1><center>المفاهيم الأساسية: كيف تتعلم الآلات؟</center></h1>

<h2><div dir="rtl">ما هو التعلم العميق؟</div></h2>

<p><div dir="rtl">هو محاولة تعليم الكمبيوتر التفكير مثل الإنسان. فكما نتعلم نحن من خلال رؤية الأشياء أو سماعها مرات عديدة، يتعلم الكمبيوتر من خلال رؤية الكثير من البيانات (إشارات، أو صور أو كلمات) مرارًا وتكرارًا. ثم يصبح النموذج (البرنامج) ماهرًا في تخمين أشياء جديدة (بشرط أن تكون العينات مشابهة وتحمل نفس الصفات الموجودة في البيانات التي تعلم منها النموذج).</div></p>

<h4><div dir="rtl"></b>مثال<b> (هل هذه دراجة هوائية؟)</div></h4>

<p><div dir="rtl">إذا أردنا تعليم/تدريب نموذج ليكون قادرا على التعرّف على صور الدراجات الهوائية، علينا تزويد النموذج بالعديد من صور الدراجات، والعديد من الصور الأخرى التي لا تعتبر دراجات هوائية، مثل صور سيارات، زهور، أشخاص، قطط، الخ.</div></p>

<p><div dir="rtl">عند تدريبنا للنموذج، علينا أن نخبره أن هذه الصور هي "دراجة هوائية" ودعنا نسمي صور غير الدراجات "أخرى". في كل صورة يراها النموذج سيحاول الإجابة على سؤالنا: ما هذه الصورة "دراجة هوائية" أم "أخرى". في حال كانت إجابته خاطئة نخبره بذلك فيقوم بتحديث معرفته. وفي حال كانت إجابته صحيحة ايضا نخبره ليعرف أن إجابته صحيحة ويقوم بتأكيد معرفته.</div></p>

<p><div dir="rtl">كلما كانت البيانات التي يراها النموذج اكثر كلما كانت معرفته وتمييزه أفضل.</div></p>

<p><div dir="rtl">في حال كان تدريبنا للمودل ناجحا وزودنا النموذج بالكم الكافي من الصور سيكون قادرا على تمييز الصور الجيد ما إذا كانت "دراجة هوائية" أو "أخرى".</div></p>

<p><div dir="rtl">عملية التنبؤ والتخمين في التعلم العميق ذات طبيعة احتمالية</div></p>

<p><div dir="rtl">مهمة النموذج هي التعرف على الصور، لكن معرفته محدودة في تسمية الدراجات فقط. في حال عرضنا عليه أي صورة غير الدراجات، قطة كانت أو سيارة، فأنه سيتعرف عليها على أنها ليست دراجة "أخرى".</div></p>


<p><div dir="rtl">لن يستطيع النموذج التعرف على القطط أو السيارات، فكل شي ضمن معرفته غير الدراجات الهوائية اسمه "أخرى".</div></p>
<p><div dir="rtl">وبالتالي عند تعرفه على أي صورة، سيقوم بتوزيع النتائج بنسبة مئوية (معرفته الكلية) على هذين الصنفين.</div></p>


<p><div dir="rtl">بما أن معرفة النموذج الكاملة 100% تتكون من نوعين فقط "دراجة هوائية" و "أخرى" فإن إجابته ستكون على الشكل التالي:</div></p>


<div dir="rtl">
    <ul>
        <li>لو عرضنا عليه صورة دراجة يجب أن يخبرنا أنها "دراجة هوائية" بنسبة 100% و "أخرى" بنسبة 0%.</li>
        <li>لو عرضنا عليه صورة طاولة يجب أن يخبرنا أنه "أخرى"  بنسبة 100% و "دراجة هوائية" بنسبة 0%.</li>
        <li>لو عرضنا عليها صورة قلم يجب أن يخبرنا أنه "أخرى"  بنسبة 100% و "دراجة هوائية" بنسبة 0%.</li>
        <li>لو عرضنا عليها صورة سيارة يجب أن يخبرنا أنها "أخرى"  بنسبة 100% و "دراجة هوائية" بنسبة 0%.</li>
    </ul>
</div>

<p><div dir="rtl">الأمثلة المذكورة سابقا، توضح كيف أن النموذج أصبح بارعا في التمييز بين الصنفين المستهدفين. في العالم الحقيقي غالبا ما تكون النسب أقل، مثلا:</div></p>

<div dir="rtl">
    <ul>
        <li>90% دراجة هوائية 10% أخرى</li>
        <li>70% دراجة هوائية 30% أخرى</li>
    </ul>
</div>

<p><div dir="rtl">هذه النتائج والنسب تعتمد على جودة النموذج والصورة المراد التعرف عليها.</div></p>

<p><div dir="rtl">لو كانت جميع صور الدراجات التي تعلم منها النموذج صور جانبية، وبعد التدريب قمنا بسؤاله عن صورة دراجة من الأمام، ممكن أن تكون إجابته "أخرى"  بنسبة 85% و "دراجة هوائية" بنسبة 15%.</div></p>
